```python
# simple Linear Regression

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.metrics import mean_squared_error
```

```python
# import the data set 

# data set url 
url = "/Users/dibaspaudel/Library/CloudStorage/OneDrive-UniversityofCanberra/SEM_3_2023/DST&S/Assignment/Assignment_1/data/"

# using pandas to read the data frame 
df = pd.read_csv(url + "zomato_df_final_data.csv")

df.drop_duplicates(inplace = True)
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
# Using simple Linear Regression 

# Y = mX + c

# where Y = Dependent Variable , X = Independent Variable, m & c = Tranning Variable

# Droping the other variables
df_model = df[['cost', 'rating_number']]


x_reg = df_model['cost']
y_reg = df_model['rating_number']

x_train , x_test, y_train, y_test = train_test_split(x_reg, y_reg, test_size = 0.2, random_state = 0)
```

```python
# see the corelation between 

# Create a scatter plot to visualize the correlation
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_model, x='cost', y='rating_number')
plt.xlabel("Restaurant Cost")
plt.ylabel("Restaurant Rating")
plt.title("Correlation Plot: Restaurant Cost vs. Restaurant Rating")
plt.show()
```

```python
x_train = np.array(x_train).reshape(-1,1)
x_train
```

```python
x_test = np.array(x_test).reshape(-1,1)
x_test
```

```python
LR = LinearRegression()
```

```python
# Train the Model

LR.fit(x_train, y_train)
```

```python
c = LR.intercept_
c
```

```python
m = LR.coef_
m
```

```python
# Prediction , where prediction (y = m * x_train + c)

y_prediction_train_1 = LR.predict(x_train)
y_prediction_train_1
```

```python
y = m* x_train + c
y.flatten()
```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, y_prediction_train_1, color = 'red')
plt.xlabel("Restaurant Cost")
plt.ylabel("Restaurant Rating")
```

```python
# Testing

y_prediction_test_1 = LR.predict(x_test) 
y_prediction_test_1
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_test, y_prediction_test_1, color = 'red')
plt.xlabel("Restaurant Cost")
plt.ylabel("Restaurant Rating")
```

```python
y_prediction_test_2 = LR.predict([[50]]) 
y_prediction_test_2
```

```python
# Calculate the MSE

mse = mean_squared_error(y_test, y_prediction_test_1)
mse
```

```python
# Model 2: Linear Regression with Gradient Descent
LR_2 = SGDRegressor(max_iter=1000, tol=1e-3)
LR_2.fit(x_train, y_train)
```

```python
y_prediction_2 = LR_2.predict(x_test)
```

```python
mse_2 = mean_squared_error(y_test, y_prediction_2)
mse_2
```

```python
print("Mean Squared Error (MSE) for Linear Regression:", mse)
print("Mean Squared Error (MSE) for Linear Regression with Gradient Descent:", mse_2)
```
